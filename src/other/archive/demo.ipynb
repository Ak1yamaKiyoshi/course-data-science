# ‚¨ÖÔ∏è Import librarias

```python
import cv2
import matplotlib.pyplot
from google.colab.patches import cv2_imshow
```

Creating sift instance

```python
sift = cv2.SIFT_create()
```

And sift matcher

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)
```

  For the example, I will use two downloaded files from copernicus named 'a.jpg' and 'b.jpg'

```python
img1 = cv2.imread('/content/a.jpg')
img2 = cv2.imread('/content/b.jpg')
```

We remove colors because they can affect how SIFT detects surfaces (points), and if there is a strong difference between them, SIFT may not detect a connection

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

Look for key points in both images

```python
keypoints_1, descriptors_1 = sift.detectAndCompute(img1, None)
keypoints_2, descriptors_2 = sift.detectAndCompute(img2, None)
```

```python
matches = bf.match(descriptors_1, descriptors_2)
matches = sorted(matches, key=lambda x:x.distance)
```

Demonstration of results

```python
img3 = cv2.drawMatches(img1, keypoints_1, img2, keypoints_2, matches[300:600], img2, flags=2 )
```

```python
cv2_imshow( img3)
```
